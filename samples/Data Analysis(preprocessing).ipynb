```python
import pandas as pd
```

```python
data = pd.read_csv('Drug data.csv')
data.head(10)
```

```python
age = pd.qcut(data['Age'],2)
data.pivot_table('Na_to_K',[age,'Sex'],'BP',aggfunc='mean')
```

```python
import numpy as np
```

```python
mapping = {'LOW':1,'NORMAL':2,'HIGH':3}
data['BP'] = data['BP'].map(mapping)
data.head()
```

```python
inverse = {v:k for k,v in mapping.items()}
data['BP'] = data['BP'].map(inverse)
data.head()
```

```python
sex_mapping = {label: idx for idx, label in enumerate(np.unique(data['Sex']))}
sex_mapping
```

```python
data['Sex'] = data['Sex'].map(sex_mapping)
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
data.iloc[:,5] = LabelEncoder().fit_transform(data.iloc[:,5])
data.head(10)
```

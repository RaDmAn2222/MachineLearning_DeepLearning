```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
data = pd.read_csv('winequality-red.csv')
data.head(10)
```

```python
columns = ['fixed acidity','residual sugar','chlorides','free sulfur dioxide','pH',]
new_data = data.drop(columns, axis=1)
new_data.head(10)
```

```python
X = new_data[['volatile acidity','citric acid','total sulfur dioxide','density','sulphates','alcohol']]
X[0:5]
```

```python
y = data[['quality']]
y[0:5]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_trainset, X_testset, y_trainset, y_testset = train_test_split(X, y, test_size=0.3, random_state=3)
```

```python
qualitytree = DecisionTreeClassifier(criterion='entropy')
qualitytree
```

```python
qualitytree.fit(X_trainset, y_trainset)
```

```python
predTree = qualitytree.predict(X_testset)
```

```python
print(predTree[0:5])
print(y_testset[0:5])
```

```python
from sklearn import metrics
import matplotlib.pyplot as plt
print('Accuracy is: ', metrics.accuracy_score(y_testset, predTree))
```
